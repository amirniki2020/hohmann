#Make  A Hohmann Transfer Orbit!

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from IPython.html import widgets
from IPython.display import display
```

```python
from poliastro.bodies import Body, Sun
from poliastro.plotting import OrbitPlotter
from hohmann import hoh_data, ss_data, create_states, display_data
```

---

##Provide The Orbital Coniditions

```python
ri = widgets.FloatText(value=1,
    description='init rad:')
rf = widgets.FloatText(value=4.5,
    description='final rad:')

display(ri)
display(rf)
```

---

###Calculate the orbits with the commands below:

```python
Star = Body(0.51*Sun.k, R=0.5*u.AU)
ssi,ssa,ssf,hoh = create_states(Star,ri,rf)
```

```python
display_data(ssi,hoh,ssf)
```

---

###Make a pretty plot of the orbits:

```python
op = OrbitPlotter()
full = op.plot(ssi)+op.plot(ssa)+op.plot(ssf)
```
