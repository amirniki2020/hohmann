#Make  A Hohmann Transfer Orbit!

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from IPython.html import widgets
from IPython.display import display
```

```python
from poliastro.bodies import Body, Sun, Earth
from poliastro.plotting import OrbitPlotter
from hohmann import (hoh_data, ss_data, create_states,
                    display_data, create_widgets)
```

---

##Provide The Orbital Coniditions

```python
bodies = {'sun masses':Sun, 'earth masses':Earth}
units = {'km':u.km, 'au':u.AU}
ddm,ms,ddu,ri,rf = create_widgets()
display(ddm,ms,ddu,ri,rf)
```

---

###Calculate the orbits with the commands below:

```python
cm = bodies[ddm.value]
unit = units[ddu.value]
Star = Body(ms.value*cm.k, R=cm.R)
ssi,ssa,ssf,hoh = create_states(Star,ri,rf,unit)
```

```python
display_data(ssi,hoh,ssf)
```

---

###Make a pretty plot of the orbits:

```python
op = OrbitPlotter()
full = op.plot(ssi)+op.plot(ssa)+op.plot(ssf)
```
